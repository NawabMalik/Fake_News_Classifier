```python
df=train.drop(['label'], axis=1)
df.head()
```

```python
Y=train['label']
Y.head()
```

```python
porter=PorterStemmer()
corpus=[]
for i in range(0,len(df['text'])):
    review=re.sub('^[a-zA-Z]', ' ', df['text'][i])
    review=review.lower()
    review=review.split()
    review=[porter.stem(word) for word in review if word not in set(stopwords.words('english'))]
    review=' '.join(review)
    corpus.append(review)
```

```python
corpus[2]
```

```python
# lets create Bag of words:
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=10000, ngram_range=(1,3))      #n_gram means combinations of features, 1 or 2 or 3:
X=cv.fit_transform(corpus)
```

```python
(X.shape), cv.get_feature_names()[:10]          #all 20000 features name are here
```

```python
cv.get_params()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y, test_size=0.2, random_state=0)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB().fit(X_train,Y_train)
model_pred=model.predict(X_test)
confusion_matrix(model_pred, Y_test)
```

```python
accuracy_score(model_pred, Y_test)
```

```python
# accuracy on test data:
model_fit_test=MultinomialNB().fit(X_test,Y_test)
model_pred_test=model_fit_test.predict(X_train)
confusion_matrix(model_pred_test, Y_train)
```

```python
accuracy_score(model_pred_test, Y_train)
```
